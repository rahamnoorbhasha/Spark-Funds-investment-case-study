## Removing Null values
### We need to find the average amount invested in each funding type. So the rows with out amount(raised_amt_usd) can be ignored

```python
# Ignoring rows with null values in raised_amount_usd column
rounds2 = rounds2[~rounds2['raised_amount_usd'].isnull()]
rounds2.info()
```

### Formatting permalink and company_permalink in companies and round2 dataframes respectively to lower case so that we will have unique format across the columns that we use for merging.

```python
rounds2_formatted = rounds2.copy()
companies_formatted = companies.copy()
rounds2_formatted['company_permalink'] = rounds2['company_permalink'].str.lower()
companies_formatted['permalink'] = companies['permalink'].str.lower()
companies_formatted.head()
```

```python
rounds2_formatted.head()
```

### Merged two dataframe to create master_frame for furthur Analysis

```python
master_frame = pd.merge(rounds2_formatted,companies_formatted,left_on='company_permalink',right_on='permalink',how='left')
master_frame
```

```python
master_frame.info()
```

### There are null values in the country code and category list, We can not remove the null values in these rows now as it may remove the rows that have valid funding type and raised amount which will impact the calculation of best funding type. ### Exracting the primary sector from category_list in master_frame and add it as column

```python
master_frame['primary_sector'] = master_frame['category_list'].str.split('|').str[0]
```

```python
master_frame
```

### Extracting mapping.csv data to  dataframe

```python
mapping = pd.read_csv('mapping.csv')
mapping
```

```python
mapping.info()
```

### There is null value in category list and this row is not required for our analysis. We have column called which is not a sector. So clean mapping to remove null values.

### In category_list column the data is missplet where 0 is there in place of na. Eg: A0lytics instead of Analytics. So need to correct.

```python
mapping = mapping[~(mapping['category_list'].isnull())].drop(columns='Blanks') #Removeing nulls from category_list

# Replacing 0 with 'na'

mapping.loc[mapping['category_list'] != 'Enterprise 2.0', 'category_list'] = mapping.category_list.str.replace("0","na")
mapping
```

### create column in mapping dataframe to tag main_sector of each category in the category list which can be used for merging prmiary_sector in master frame and main_sector in mapping dataframe

```python
mapping['main_sector'] = (mapping == 1).idxmax(axis=1)
mapping
```

### Merging master data frame with mapping data frame to add main_sector to the master_frame.

```python
master_frame = pd.merge(master_frame,mapping[['category_list','main_sector']],left_on='primary_sector',right_on='category_list',how='left')
master_frame = master_frame.drop(columns=['category_list_y'])
master_frame = master_frame.rename(columns = {'category_list_x':'category_list'})
master_frame
```

```python
master_frame.info()
```

### There are null values in the main_sector, As we are doing sector wise analysis we can ignore the rows with null values in this field. So removing rows with null values in main_sector

```python
#ignoring rows with null values in main_sector
master_frame = master_frame[~master_frame['main_sector'].isnull()].copy()
master_frame.info()
master_frame
```

### Merged Dataframe with primary sector and mainsector

```python

merged_df = master_frame[['primary_sector','main_sector']]
merged_df = merged_df.drop_duplicates('primary_sector')
merged_df
```

### Creating Datasets for three countries that we found in previous steps for funding round type - Venture and in investment range 5-15M

```python
D1 = master_frame[(master_frame['country_code'] == 'USA') & (master_frame['funding_round_type'] == 'venture') & 
              (master_frame['raised_amount_usd'] > 5000000) & (master_frame['raised_amount_usd'] < 15000000)].copy()
D2 = master_frame[(master_frame['country_code'] == 'GBR') & (master_frame['funding_round_type'] == 'venture') & 
              (master_frame['raised_amount_usd'] > 5000000) & (master_frame['raised_amount_usd'] < 15000000)].copy()
D3 = master_frame[(master_frame['country_code'] == 'IND') & (master_frame['funding_round_type'] == 'venture') & 
              (master_frame['raised_amount_usd'] > 5000000) & (master_frame['raised_amount_usd'] < 15000000)].copy()
```

### From these data sets we need to do sector wise analysis 

```python
#D1.groupby('main_sector')['raised_amount_usd'].sum() # Total Sum for different mainsectors
#D1.groupby('main_sector')['raised_amount_usd'].count() # Total count for different mainsectors

# By Transposing above dataframes we can create sum and count columns in D1

D1['Totval'] = D1.groupby('main_sector')['raised_amount_usd'].transform('sum')
D1['Totcount'] = D1.groupby('main_sector')['raised_amount_usd'].transform('count')
D1
```

```python
# Adding the sum and count as columns to D2
D2['Totval'] = D2.groupby('main_sector')['raised_amount_usd'].transform('sum')
D2['Totcount'] = D2.groupby('main_sector')['raised_amount_usd'].transform('count')
D2
```

```python
# Adding the sum and count as columns to D3
D3['Totval'] = D3.groupby('main_sector')['raised_amount_usd'].transform('sum')
D3['Totcount'] = D3.groupby('main_sector')['raised_amount_usd'].transform('count')
D3
```

### Extracting solutions from D1, D2, D3

```python
print('Total number of investments :' + str(D1[['main_sector','Totval','Totcount']].drop_duplicates()['Totcount'].sum()))
print('Total amount of investments :' + str(D1[['main_sector','Totval','Totcount']].drop_duplicates()['Totval'].sum()))
print(D1[['main_sector','Totval','Totcount']].drop_duplicates().sort_values('Totcount',ascending=False))

```

```python
# Companies that invested more in top and second top main_sector
print(D1.loc[D1[D1['main_sector']=='Others']['raised_amount_usd'].idxmax()]['name'])
print(D1.loc[D1[D1['main_sector']== 'Social, Finance, Analytics, Advertising']['raised_amount_usd'].idxmax()]['name'])
```
